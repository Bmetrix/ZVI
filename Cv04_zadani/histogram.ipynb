```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.cvtColor(cv2.imread("kytka.png", cv2.IMREAD_COLOR) , cv2.COLOR_BGR2RGB)
img_gray = cv2.imread("kytka.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray') 
```

```python

'''
calculate histogram
'''
hist = cv2.calcHist([img_gray], [0], None, [256], [0,255])

print (hist.shape)
    
plt.plot(hist,color = 'b')
plt.xlim([0,256])
plt.show()
```
